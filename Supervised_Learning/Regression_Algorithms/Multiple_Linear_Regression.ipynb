```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Example dataset
data = {
    'Experience': [1, 2, 3, 4, 5],
    'Test_Score': [80, 85, 88, 90, 95],
    'Interview_Score': [70, 75, 80, 85, 90],
    'Salary': [30000, 35000, 40000, 45000, 50000]
}

df = pd.DataFrame(data)
X = df[['Experience', 'Test_Score', 'Interview_Score']]
y = df['Salary']

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)

# Predict
new_data = pd.DataFrame([[6, 92, 88]], columns=['Experience', 'Test_Score', 'Interview_Score'])
predicted = model.predict(new_data)
print(f"Predicted Salary: â‚¹{predicted[0]:,.2f}")

print("Model Coefficients:",model.coef_)
print("Model Intercept:", model.intercept_)

import numpy as np
import matplotlib.pyplot as plt

# Fix test_score and interview_score
fixed_test_score = 90
fixed_interview_score = 85

experience_range = np.linspace(1, 6, 100)
X_plot = pd.DataFrame({
    'Experience': experience_range,
    'Test_Score': [fixed_test_score]*100,
    'Interview_Score': [fixed_interview_score]*100
})

y_plot = model.predict(X_plot)

plt.figure(figsize=(8, 5))
plt.plot(experience_range, y_plot, color='red', label='Prediction')
plt.scatter(df['Experience'], y, color='blue', label='Actual')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience\n(Test Score=90, Interview Score=85)')
plt.grid(True)
plt.legend()
plt.show()

# Display the pairplot Heatmap
import seaborn as sns
sns.pairplot(df)
```
